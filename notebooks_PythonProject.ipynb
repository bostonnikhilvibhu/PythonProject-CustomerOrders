```python
##store starts

customer_name = ['Nikhil','Amit','Narendra','Kalpana','Akshar']

orders = [
          ('Nikhil','Washing Machine',100,'Home Essentials'),
          ('Nikhil','Laptop',500,'Electronics'),
          ('Amit','Mobile',200,'Electronics'),
          ('Amit', 'Laptop',500,'Electronics'),
          ('Amit', 'SmartWatch',300,'Electronics'),
          ('Narendra','T-shirt',40,'Clothing'),
          ('Kalpana','TV',80,'Electronics'),
          ('Kalpana','Juicer-Mixer',10,'Home Essentials'),
          ('Akshar','Suit',40,'Clothing'),
          ('Akshar','Battery',5,'Electronics')
]

## store ends

#Declaration for empty data strctures
customer_ordered = {}
distinct_products = {}
distinct_categories = {}
map_product_category = {}

##classify product by category starts

customer_ordered = {name:[order[1] for order in orders if order[0]==name] for name in customer_name}
print("customer_ordered ",customer_ordered)

distinct_products = {order[1] for order in orders}
print("distinct product info ", distinct_products)

distinct_categories = {order[3] for order in orders}
print("distinct categories info   " ,distinct_categories)

map_product_category = {product:category for name,product,price,category in orders} ## unpacking tuples
map_product_category = dict(sorted(map_product_category.items(),key = lambda x : x[1])) ## sort product map by category

print("Product Category " ,map_product_category)                        
          
## classify product by category ends 

## Analyze customer order starts
customer_total = {}
customer_value={}
customer_total = {name:sum([order[2] for order in orders if order[0]==name]) for name in customer_name}
#print(customer_total)

customer_value = {}
customer_value= {name: "HighValue Customer" if value>=100 
                 else "ModerateValue Customer" if (value >= 50 and value < 100)
                 else "LowValue Customer" for name,value in customer_total.items()}

#print(customer_value)

## Analyze customer order ends

##Generate business insigts starts
total_revenue_per_productcategory = {}
top_three_highest_customer_spending = dict(sorted(customer_total.items(),key = lambda x :x[1],reverse =True)[:3]);
print("top_three_highest_customer_spending ", top_three_highest_customer_spending)

total_revenue_per_productcategory = dict(sorted({category:sum([order[2] for order in orders if order[3]== category])
                                              for category in distinct_categories}.items(),key =lambda x: x[1],reverse=True))


customer_purchased_electronics = list(set(order[0] for order in orders if order[3] == 'Electronics' ))
print("customer_purchased_electrnocis -- ",customer_purchased_electronics)
customer_purchased_clothing = list(set(order[0] for order in orders if order[3] == 'Clothing'))
print("customer purchased clothing -- ", customer_purchased_clothing)
customer_purchased_electronics_clothing = set(customer_purchased_electronics) & set(customer_purchased_clothing)

##

## Generate business insights ends

## Organize and display starts

for (key1,value1),(key2,value2) in zip(customer_ordered.items(),customer_value.items()):
    print("Customer Name    ", key1)
    print("Customer Ordered ",value1)
    print("Customer value   ", value2)
    print("-----------------------")
    
## Organize and display ends

customer_categories = {
                        name:{order[3] for order in orders if order[0] == name} for name in customer_name}

#print(customer_categories)

#for i in customer_categories.items():
 #   print(i, type(i), type(i[0]),type(i[1]))

customer_multiple_categories = [name for name,category in customer_categories.items() if len(category)>1]
print("customer bought Multiple Categories  ", customer_multiple_categories)

print("customer_purchased_electronics_clothing ", customer_purchased_electronics_clothing)


## organize and display ends

## Actions - Dict methods

customer_details_dict = {name:[order for order in orders if order[0]== name] for name in customer_name}
#print("customer_details ", customer_details_dict)

print("Get Nikhil details ", customer_details_dict.get("Nikhil"))

print(customer_details_dict.get("Unknown") ,"--- Even though it is Unknown customer, it should not fail since get mehtod is used.")


## Trends

category_sales_info = {
    category: sum(price for _, _, price, cat in orders if cat == category)
    for category in distinct_categories
}

print(category_sales_info)
print("total_revenue_per_productcategory  ", total_revenue_per_productcategory)
print("max sales category ", max(
    category_sales_info.keys(),
    key=lambda k: category_sales_info[k]
))

print("min sales category ", min(
    category_sales_info.keys(),
    key=lambda k: category_sales_info[k]
))

total_sales = sum(category_sales_info.values())

print("total_sales_info ", total_sales)

category_percent_info = {
    cat: round((value / total_sales) * 100, 2)
    for cat, value in category_sales_info.items()
}

print(category_percent_info)
```
